```python
import numpy as np

def get_option():
    """
    Display menu options
    """
    
    print("PSU Library Insight Software\n\nAvalible Services:\n")
    print(" 1) Check Avalability\n 2) Most popular books\n 3) Borrow ratio\n 4) Usage ratio\n 5) Pending Fines\n 6) Exit") 
    option = int(input("Please enter desired service: "))
    
    if option in [1,2,3,4,5,6]:
        return option
    else:
        print("Invalid option!")
        return get_option()
    
def load_book_list(file_name):
    """
    Read book list file
    """
    book_id = []
    names = []
    copies = []
    restricted = []
    
    copies_left = []
    total_borrowed_days = []
    largest_no_copies_borrowed = []
    
    
    booklist = open(file_name, "r")
    
    
    line = booklist.readline()
    i = 0
    while line != "":
        line = line.rstrip("\n")
        m = line.split("#")
    
        names.append (m[0])
        copies.append (int(m[1]))
        restricted.append (m[2])
        book_id.append(i)
        i += 1 
        copies_left.append(int(m[1]))
        total_borrowed_days.append(0)
        largest_no_copies_borrowed.append(0)
        
        line = booklist.readline()
        
       
        
    print(len(names), " books loaded!")
        
    return [book_id, names, copies, restricted, copies_left, total_borrowed_days, largest_no_copies_borrowed]

def get_index_max_number(a):
    indx = 0
    max_value = 0
    for i in range(len(a)):
        if a[i] > max_value:
            max_value = a[i]
            indx = i
    return indx, max_value

def get_book_id(book_name, books):
    """
    Get book record for a given book name
    """
    i = 0 
    for bk in books[1]:
        if bk == book_name:       
            return books[0][i]
        i = i+1
    print(book_name, " not found")
    return -1

def get_borrow_ratio(books):
    br = []
    
    for i in range(len(books[6])):
        br.append(books[6][i]/books[2][i])
    return br

# TODO: we need to sort in descending order
def sort_books_by_borrow_ratio(br, names):
    ind = np.argsort(br)
    ordered_names = [names[i] for i in ind]
    return ordered_names

def load_library_logs(logs_file, books):
    """
    Read library log file
    """
    borrow_day = []
    borrow_sn = []
    borrow_bn = []
    borrow_dbf = []   # no of days the books is borrowed for
    borrow_due_day = []
    borrow_retricted = []
    
    return_day = []
    return_sn = []
    return_bn = []
    return_charged = []
    
    add_day = []
    add_bn = []
    
    pay_day = []
    pay_sn = []
    pay_amount = []
    
    
    charge_day = []  
    charge_amount = []
    
    
    
    i = 1
    reader = open(logs_file, "r")
    
    
    line = reader.readline()

    while line != "":
        
        line = line.rstrip("\n")
        m = line.split("#")

        if m[0] == "B":
            # read borrow log
            borrow_day.append(m[1])
            borrow_sn.append(m[2])
            borrow_bn.append(m[3])
            borrow_dbf.append(m[4])
            borrow_due_day.append( str(int(m[1]) + int(m[4]))) # if you borrow on day 1 for three days, the due date is 4
            bkid = get_book_id(m[3], books)
            borrow_retricted.append(books[3][bkid]) 
            
            # update book profile
            books[4][bkid] = books[4][bkid] - 1  # decrement no of copies left
            books[5][bkid] = books[5][bkid] + int(m[4])
            current_no_copies_borrowed = books[2][bkid] - books[4][bkid]
            if current_no_copies_borrowed > books[6][bkid]:
                books[6][bkid] = current_no_copies_borrowed

        elif m[0] == "R":
            return_day.append(m[1])
            charge_day.append(m[1])
            return_sn.append(m[2])
            return_bn.append(m[3])
            dd, rst = get_borrow_record(m[2], m[3], borrow_day, borrow_sn, borrow_bn, borrow_due_day, borrow_retricted)
            diff = int(m[1]) - int(dd)
            if diff <= 0:
                return_charged.append("0") #TODO turen into number
                charge_amount.append("0")
            else:
                if rst == "TRUE":
                    return_charged.append(str(diff*5))
                    charge_amount.append(str(diff*5))
                else:
                    return_charged.append(str(diff))
                    charge_amount.append(str(diff))
            # update book profile
            bkid = get_book_id(m[3], books)
            books[4][bkid] += 1
                    

        elif m[0] == "A":
            # Does the book exist
            bkid = get_book_id(m[2], books)
            if bkid != - 1:
                books[2][bkid] += 1
            else:   # add a new book
                bkid = len(books[0])
                books[0].append(bkid)
                books[1].append(m[2])
                books[2].append(1)
                books[3].append("FALSE")  # Note: add commaned sould specify whether the new books is restricted or not
                books[4].append(1)
                books[5].append(0)
                books[6].append(0)
                
 
        elif m[0] == "P":
            pay_day.append(m[1])
            pay_sn.append(m[2])
            pay_amount.append(m[3])

        else:  # end of logs 
            if len(m) > 1:
                print("log file error in line ", i)
            
        i = i+1
        line = reader.readline()
        
    borrow_logs = [borrow_day, borrow_sn, borrow_bn, borrow_due_day, borrow_retricted]
    return_logs = [return_day, return_sn, return_bn, return_charged]
    add_logs = [add_day, add_bn]
    pay_logs = [pay_day, pay_sn, pay_amount]
    charge_logs = [charge_day, charge_amount]
    
    return [borrow_logs, return_logs, add_logs, pay_logs, charge_logs]

def get_borrow_record(student_name, book_name, borrow_day, borrow_sn, borrow_bn, borrow_due_day, borrow_retricted):
    """
    Get the most recent borrow log for a given student and book
    """
    
    query = student_name + book_name
    
    index = -1
    for i in range(len(borrow_sn)):
        if query == (borrow_sn[i] + borrow_bn[i]):
            index = i
    if index == -1:
        print("Error ", student_nabme, " never borrowed ", book_name)
    
    return  borrow_due_day[index], borrow_retricted[index]
            
#TODO: Never used. Work in progress. consider creating student profile instead.
def create_fine_profile(books, logs):
    """
    <return_day>#<student_name>#<book_name>#<fine>
    """
    r_logs = logs[1]
    for item in logs:
        r_day = item[0]  # 
        r_sn = item[1]
        r_bn = item[2]
    return 0
        
    
    
    
def get_num_borrowed_books(std_name, logs):
    """
    Determine how many books are borrowed by a given student
    """
    borrow_records = logs[0]
    num_borrowed_books = 0
    for name in borrow_records[1]:
        if name == std_name:
            num_borrowed_books = num_borrowed_books + 1
        
    returned_records = logs[1]
    num_returned_books = 0
    for name in returned_records[1]:
        if name == std_name:
            num_returned_books = num_returned_books + 1
            
    return num_borrowed_books - num_returned_books
 
def get_owned_charges(std_name, logs):

    returned_records = logs[1]
    
    
    names = returned_records[1]
    charges = returned_records[3]
    
    total_charges = 0
    for i in range(len(names)):
        if std_name == names[i]:
            total_charges = total_charges + float(charges[i])
            
   
    
    payment_records = logs[3]
    names = payment_records[1]
    amounts = payment_records[2]
    
    total_payments = 0
    for i in range(len(names)):
        if std_name == names[i]:
            total_payments = total_payments + float(amounts[i])
            
    return total_charges - total_payments
    

def can_borrow(std_name, book_name, num_days, books, logs):
    """
    Can a specified student borrow a given book?
    """
    bkid = get_book_id(book_name, books)
    n = books[1][bkid]
    c = books[2][bkid]
    r = books[3][bkid]
    
    if c ==  0:
        print("Book not available")
        return False
    if r == "TRUE" and num_days > 7:
        print("Restricted books can't be borrowed for more than 7 days")
        return False
    if  num_days > 28:
        print("Book can't be borrowed for more than 28 days")
        return False
    
    if get_num_borrowed_books(std_name, logs) >= 3:
        print("Student can't borrow more than 3 books")
        return False
    
    #  check for pending fines
    if get_owned_charges(std_name,logs) > 0:
        print("Student has unpaid fines")
        return False
        
    return True
    

def get_pending_payments(day, pay_logs, charge_logs):
    total_pay = 0.0
    total_charge = 0.0
    """
    print("charges")
    print(charge_logs[0])
    print(charge_logs[1])
    print("payments")
    print(pay_logs[0])
    print(pay_logs[2])
    """
    
    for i in range(len(pay_logs[0])):
        if int(pay_logs[0][i]) <= day:
            total_pay += float (pay_logs[2][i])
        else:
            break
            
    for i in range(len(charge_logs[0])):
        if int(charge_logs[0][i]) <= day:
            total_charge += float (charge_logs[1][i])
        else:
            break
                        
    return (total_charge - total_pay)
            
           
```

```python
book_list_file = "./Booklist.txt"
logs_file = "././Librarylog.txt"
```

```python
books = load_book_list(book_list_file)
logs = load_library_logs(logs_file, books)
```

## Main code

```python
cont = True
while cont:
    opt = get_option()
    if opt == 1:
        book_str = input("Enter <student_name>#<book_name>#<days>")
        a = book_str.split("#")
        if can_borrow(a[0],a[1],int(a[2]), books, logs):
            print("Yes")
        else:
            print("No")
        
    elif opt == 2:
        indx, val = get_index_max_number(books[5])
        print("The most borrowed book is ", books[1][indx], ". Total days borrowed is ", val)
    elif opt == 3:
        br = get_borrow_ratio(books)
        indx, val = get_index_max_number(br)
        print("The book with the highest borrow ratio is:  ", books[1][indx], ". Ratio is ", val)
        
    elif opt == 4:
        br = get_borrow_ratio(books)
        mbb = sort_books_by_borrow_ratio(br, books[1])
        for i in range(-1, -len(mbb)-1, -1):
            print(mbb[i])
        
    elif opt == 5:
        day = int(input("Enter a day: "))
        total_pending_fine = get_pending_payments(day, logs[3], logs[4])
        print("Total pending fine: ", total_pending_fine, "\n")
        
    if opt == 6:
        cont = False
        
```
